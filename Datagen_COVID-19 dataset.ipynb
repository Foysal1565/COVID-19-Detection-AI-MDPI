```python
import pandas as pd
import os
import shutil
```

```python
#COVID-19 positive image data


FILE_PATH = 'chestxray/metadata.csv'
IMAGES_PATH = 'chestxray/images'
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR = 'Dataset/Covid'

if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print('folder created')
```

```python
cnt = 0
for (i, row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19"and row["view"]=="PA" :
        filename= row["filename"]
        image_path = os.path.join(IMAGES_PATH,filename)
        image_copy_path = os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        print("Moving image", cnt)
        cnt += 1
#print(cnt)
```

```python
#Sampling Images from Kaggle

import random

KAGGLE_FILE_PATH = "chestxray_kaggle/train/NORMAL"
TARGET_NORMAL_DIR = "Dataset/Normal"
```

```python
image_names=os.listdir(KAGGLE_FILE_PATH)
random.shuffle(image_names)

```

```python
for i in range(1000):
    images = image_names[i]
    image_path = os.path.join(KAGGLE_FILE_PATH,images)
    target_path = os.path.join(TARGET_NORMAL_DIR,images)
    shutil.copy2(image_path,target_path)
    print ("copying normal chest Xrays", i)
    
    
```
